```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
from scipy import signal
import soundfile
import numpy as np
import os

SR = 44100

# Update the path to your IR here
ir_verb, sr = librosa.load('./dataset/gill-heads-mine/b-format/mine_site1_1way_bformat.wav',sr=SR)
ir_verb = librosa.resample(ir_verb, sr, 44100)
ir_verb = np.multiply(ir_verb, 1.0/np.max(ir_verb))


for root, dirs, files in os.walk("./dataset/guitarset"):
    for name in files:
        audio, sr = librosa.load(os.path.join(root,name),sr=SR)
        filtered = signal.convolve(audio, ir_verb, method='fft')
        soundfile.write(os.path.join('./dataset/mine_guitarset/', name), filtered, SR)
        print("Generated file for {}".format(os.path.join(root, name)))
```
